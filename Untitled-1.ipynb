```python
labels = gpd.read_file('data/attributes.json')
```

```python
labels['street_edge_id'].nunique()
```

```python
#Single out curb ramp labels
curb_ramp = labels.loc[labels['label_type'] == 'CurbRamp']
curb_ramp
```

```python
#Weighted count based on severity for curb ramps
dict1 ={ 1.0:1.0, 2.0:0.8, 3.0: 0.6, 4.0:0.4, 5.0:0.2}
curb_ramp = curb_ramp.replace({"severity": dict1})
```

```python
#Weighted count based on severity for curb ramps
dict1 ={ 1.0:1.0, 2.0:0.8, 3.0: 0.6, 4.0:0.4, 5.0:0.2}
curb_ramp = curb_ramp.replace({"severity": dict1})
```

```python
#Single out other labels
label_others = labels[labels.label_type != 'CurbRamp']
```

```python
#Combine two dfs
df= pd.concat([curb_ramp, label_others])
```

```python
df
```

```python
#Group by sidewalk segments
df_grouped = df.groupby(by=["street_edge_id"]).sum()
df_grouped.head()
```

```python
def sigmoid(x):
  return 1 / (1 + math.exp(-x))
```

```python
df_grouped['score_sigmoid'] = df_grouped['severity'].apply(sigmoid)
```

```python
a = sns.histplot(data=df_grouped, x="score_sigmoid", color="#47B5B7")
sns.despine(left=True,bottom=True)
```

```python
df_grouped.reset_index(inplace=True)
```

```python
#select street_edge_id and score_sigmoid
sigmoid_score = df_grouped[['street_edge_id', 'score_sigmoid']]
```

```python
sigmoid_score
```

```python
#merge streets and sigmoid_score on street_edge_id
streets = streets.merge(sigmoid_score, on='street_edge_id', how='left')
streets
```

```python
streets.explore(column ='score_sigmoid')
```
